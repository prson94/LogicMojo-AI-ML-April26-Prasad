#Slicing

```python
my_list = [7, 8, 9, 6, 3]
sliced_list = my_list[0:4]
print(sliced_list)
sliced_list = my_list[0:3:2]
print(sliced_list)
sliced_list = my_list[::2]
print(sliced_list)
sliced_list = my_list[::-1]
print(sliced_list)
sliced_list = my_list[3::-1]
print(sliced_list)
```

#Condition Execution

```python
'''
Question 1 -
1. Take an integer as input from the user.
2. Calculate cube of the input.
3. If result > 50 print "Massive Number" else print "Small Number".
4. After this, print "EXECUTION COMPLETE!"
'''
```

```python
x = int(input("Please input a number: "))
result = x ** 3;
if(result > 50):
  print("Massive Number")
else:
  print("Small Number")
print("Execution Complete")
```
